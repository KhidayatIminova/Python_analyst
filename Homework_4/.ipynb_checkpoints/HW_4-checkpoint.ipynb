## Условие 3: 
### 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.scatter(df['price'], df['sqft_living'])
plt.title('Взаимосвязь между стоимостью и жилой площадью (scatter)')
plt.xlabel('Стоимость, млн.у.е.')
plt.ylabel('Жилая площадь, кв.фут.')
plt.show()
```

*В основном недвижимость представлена домами площадью до 6000 кв. футов и стоимостью до 2,3 млн у.е. Также наблюдаются выбросы*

```python
sns.jointplot(x=df['price'], y=df['sqft_living15'], kind='reg')
plt.title('Взаимосвязь между стоимостью и жилой площадью')
plt.xlabel('Стоимость, млн.у.е.')
plt.ylim(0, 8000)
plt.ylabel('Жилая площадь, кв.фут.')
plt.show()
```

*Как видно из графика существует прямая зависимость между жилой плошадью объекта недвижимости и его стоимостью*

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='kde')
plt.title('Взаимосвязь между стоимостью и рейтингом')
plt.xlabel('Рейтинг')
plt.ylabel('Стоимость, млн.у.е.')
plt.show()
```

*Наибольшее сосредоточение наблюдается наблюдается у домов с рейтингом 6, 7, 8 и стоимостью в диапазоне 400 - 700 тыс.у.е.*

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
plt.figure(figsize=(20, 20))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Корреляционная матрица');
```

*Признаки, коррелируемые со стоимостью (price):*
- *batrooms 0.5 (заметная линейная зависимость)*
- *sqft_living 0.7 (высокая линейная зависимость)*
- *grade 0.7 (высокая линейная зависимость)*
- *sqft_above 0.6 (заметная линейная зависимость)*
- *sqft_living15 0.6 (заметная линейная зависимость)*

```python
sns.boxplot(x=df['price'], whis=1.5)
```

*Основное количество домов имеет стоимость в диапазоне 400-700 тысяч, но также наблюдаются выбросы*
